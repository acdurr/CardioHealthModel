```python
def uniqueValues(col):
    return list(df[col].unique())

for col in df.select_dtypes('object').columns:
    print(f'{col}\n{uniqueValues(col)}\n')
```

```python
# Drop null values from all columns
df.dropna(subset=["General Health"], inplace=True)
df.dropna(subset=["Checkup"], inplace=True)
df.dropna(subset=["Skin Cancer"], inplace=True)
df.dropna(subset=["Other Cancer"], inplace=True)
df.dropna(subset=["Depression"], inplace=True)
df.dropna(subset=["Diabetes"], inplace=True)
df.dropna(subset=["Smoking History"], inplace=True)
df.dropna(subset=["Exercise"], inplace=True)
df.dropna(subset=["Heart Disease"], inplace=True)
df.dropna(subset=["Arthritis"], inplace=True)
df.dropna(subset=["Age"], inplace=True)
df.dropna(subset=["Height (cm)"], inplace=True)
df.dropna(subset=["Weight (g)"], inplace=True)
df.dropna(subset=["BMI"], inplace=True)

#Drop _BMI5 as it is redundant
df.drop('_BMI5', axis=1, inplace=True)
```

```python
df = df.rename(columns={'Weight (g)': 'Weight (kg)'})
```

```python
df['Weight (g)'] = df['Weight (g)'].div(100).round(2)
```

```python
df.to_csv('cleaned.csv', index=False)
```
